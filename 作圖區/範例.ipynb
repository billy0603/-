# 挖洞圓餅圖

```python
#-*- coding: utf-8 -*-
from matplotlib import pyplot as plt
plt.rcParams['font.family']='DFKai-SB' #解决中文乱码

plt.figure(figsize=(9,11)) #調整圖形大小
labels=list(data['bitn_nm'])#定義標籤
sizes=list(data['count'])#使用的值，使用總團數
colors = ['red','yellowgreen','lightskyblue','yellow','blue','orange','blueviolet','darkgreen'] #每塊顏色
#explode = (0,0,0,0.05,0.05,0,0,0) #將某塊分割出來，數值越大分割間距越大

my_circle=plt.Circle( (0,0), 0.5, color='white')

patches,texts1,texts2 = plt.pie(sizes,
                 # explode=(0,0,0,0.05,0.05,0,0,0),#分割出來有陰影
                  labels=labels,
                  colors=colors,
                  autopct = '%3.2f%%', #數值保留至小數點第幾位
                  shadow = True, #無陰影設定
                  startangle =20, #逆時針開始角度設置
                  pctdistance = 0.8,#數據距離圓心半徑倍數的距離
                  textprops={'fontsize': 18}) #label大小


p=plt.gcf()
p.gca().add_artist(my_circle)
#patches饼图的返回值，texts1饼图外label的文本，texts2饼图内部的文本
# x，y軸刻度一樣，保證為圓形
plt.axis('equal')
legend=plt.legend(loc = 'lower right',
           #title="地區",
           bbox_to_anchor=(0.015, 0.015),
           fontsize = 'x-large')

#設置legend title字體
legend.set_title("地區", prop = {'size':25})

t='開團比率'
plt.text(-0.25, -0.07, t, ha='left', wrap=True,size=30,color='red')



#設定圖內文字資訊
plt.setp(texts2, size=15,color='black')
#設定標題
#plt.title("開團筆數",size=30,color='black')
plt.show()
```

# 圓餅圖

```python
#-*- coding: utf-8 -*-
from matplotlib import pyplot as plt
plt.rcParams['font.family']='DFKai-SB' #解决中文乱码

plt.figure(figsize=(9,11)) #調整圖形大小
labels=list(data['bitn_nm'])#定義標籤
sizes=list(data['count'])#使用的值，使用總團數
colors = ['red','yellowgreen','lightskyblue','yellow','blue','orange','blueviolet','darkgreen'] #每塊顏色

patches,texts1,texts2 = plt.pie(sizes,
                  explode=(0.05,0,0,0,0,0,0,0),#將某塊分割出來，數值越大分割間距越大
                  labels=labels,
                  colors=colors,
                  autopct = '%3.2f%%', #數值保留至小數點第幾位
                  shadow = True, #無陰影設定
                  startangle =35, #逆時針開始角度設置
                  pctdistance = 0.8,
                  textprops={'fontsize': 18}) #label大小
#patches饼图的返回值，texts1饼图外label的文本，texts2饼图内部的文本
# x，y軸刻度一樣，保證為圓形
plt.axis('equal')
legend=plt.legend(loc = 'lower right',
           title="地區",
           bbox_to_anchor=(0.015, 0.015),
           fontsize = '15')

#設置legend title字體
legend.set_title("地區", prop = {'size':'x-large'})


#設定圖內文字資訊
plt.setp(texts2, size=20,color='b')
#設定標題
plt.title("開團筆數",size=30,color='black')
plt.show()
```

# 單月份成團率

```python
data=pd.read_csv("../東北亞/東北亞-月份成團率.csv",engine='python',sep = ',',encoding = 'utf-8')
```

```python
data['ratio']=round(data['ratio']*100,1)
```

```python
data
```

```python
beg_month=data['beg_month']
ratio=data['ratio']
go=data['count']
```

```python
import matplotlib.ticker as mtick  

plt.rcParams['font.family']='DFKai-SB' #解决中文乱码

#設定畫布資訊
fig = plt.figure(figsize=(15,9))  
ax1 = fig.add_subplot(111)
ax1.grid(axis=' ')#不顯示格線


#y軸的小數點
fmt='%.1f%%'
yticks = mtick.FormatStrFormatter(fmt)

#ax1資訊-折線圖
ax1.plot(beg_month, ratio, 'r-D',label=u'出團率')
ax1.yaxis.set_major_formatter(yticks)
ax1.legend(loc=0,fontsize=20)

#ax1資訊-數值顯示
for i,(_x,_y) in enumerate(zip(beg_month, ratio)):  
    plt.text(_x,_y+2,str(ratio[i])+'%',color='black',fontsize=15) 

#重點要新增    
ax2 = ax1.twinx()


#ax2資訊-直方圖
ax2.bar(beg_month,go,alpha=0.3,label=u'出團數',color=['b']) 
ax2.legend(loc=2,fontsize=20)
for xx, yy in zip(beg_month,go):
    plt.text(xx, yy+0.3, str(yy), ha='center',fontsize=15)


#X軸
plt.xticks(fontsize=18)
ax1.set_xlabel('月份', fontsize = 20,color='green')
ax1.set_xticks(range(0,13), minor=False)
ax1.tick_params(axis="x", labelsize=20)


#Y軸
ax1.set_ylabel('成團率(%)', fontsize=20, color='green')
ax1.set_ylim([0, 80])
ax1.tick_params(axis="y", labelsize=18)

ax2.set_ylabel('出團數', fontsize=20, color='green')
ax2.set_ylim([0, 1000])  
ax2.tick_params(axis="y", labelsize=18)



plt.title("各月份成團率及開團數量統計表",fontsize='30')
plt.show()
```

# 各地區分月份

```python
data=pd.read_csv("../東北亞/東北亞分地區月份.csv",engine='python',sep = ',',encoding = 'utf-8')
```

```python
data.head(1)
```

```python
data2=data[['beg_month','bitn_nm','target']]

data2=data2.loc[~(data2['target'].isnull())]
data2['bitn_nm']=data2['bitn_nm'].apply(lambda x:x.split('★高雄真情旅遊★')[1])
data2.head()
```

```python
data2=data2.groupby(['bitn_nm','beg_month'])['target'].agg(['sum','count']).reset_index()

data2['ratio']=data2['sum']/data2['count']

data2['ratio']=round(data2['ratio']*100,2)
data2.head()
```

```python
x1=data2.loc[data2['bitn_nm']=='九州系列']['beg_month']
y1=data2.loc[data2['bitn_nm']=='九州系列']['ratio']
x2=data2.loc[data2['bitn_nm']=='北海道系列']['beg_month']
y2=data2.loc[data2['bitn_nm']=='北海道系列']['ratio']
x3=data2.loc[data2['bitn_nm']=='日本系列']['beg_month']
y3=data2.loc[data2['bitn_nm']=='日本系列']['ratio']
x4=data2.loc[data2['bitn_nm']=='東北系列']['beg_month']
y4=data2.loc[data2['bitn_nm']=='東北系列']['ratio']
x5=data2.loc[data2['bitn_nm']=='琉球系列']['beg_month']
y5=data2.loc[data2['bitn_nm']=='琉球系列']['ratio']
x6=data2.loc[data2['bitn_nm']=='立山系列']['beg_month']
y6=data2.loc[data2['bitn_nm']=='立山系列']['ratio']
x7=data2.loc[data2['bitn_nm']=='關東系列']['beg_month']
y7=data2.loc[data2['bitn_nm']=='關東系列']['ratio']
x8=data2.loc[data2['bitn_nm']=='關西系列']['beg_month']
y8=data2.loc[data2['bitn_nm']=='關西系列']['ratio']
```

```python
fig = plt.figure(figsize=(11,9))
s=600
alpha=0.7

ax = plt.subplot()
ax.grid(axis='')
p1=ax.scatter(x1, y1,s=s, alpha=alpha)# 改变颜色
p2=ax.scatter(x2, y2,s=s, c='green', alpha=alpha) 
p3=ax.scatter(x3, y3,s=s, c='red', alpha=alpha)
p4=ax.scatter(x4, y4,s=s, c='yellow', alpha=alpha)
p5=ax.scatter(x5, y5,s=s, c='blue', alpha=alpha)
p6=ax.scatter(x6, y6,s=s, c='orange', alpha=alpha)
p7=ax.scatter(x7, y7,s=s, c='blueviolet', alpha=alpha)
p8=ax.scatter(x8, y8,s=s, c='lightskyblue', alpha=alpha)

#標籤
legend=ax.legend(handles=[p1,p2,p3,p4,p5,p6,p7,p8],
          labels=['九州系列', '北海道系列', '日本系列', '東北系列', '琉球系列', '立山系列', '關東系列', '關西系列'],
          loc='upper right',
          fontsize=20,
          bbox_to_anchor=(1.4, 1.015),
          title="地區",) 

legend.set_title("地區", prop = {'size':30})

#X軸
plt.xlabel('月份', fontsize=20, color='green')
plt.xticks(range(0,13),fontsize=18)
plt.xlim([0, 13])



#Y軸
plt.ylabel('單地區成團比率(%)', fontsize=20, color='green')
plt.yticks(fontsize=18)




plt.title("成團比率",size=30,color='black')   

plt.show()
 
```

# 單地區泡泡圖

```python
x1=data2.loc[data2['bitn_nm']=='九州系列']['beg_month']
y1=data2.loc[data2['bitn_nm']=='九州系列']['ratio']
z1=data2.loc[data2['bitn_nm']=='九州系列']['count']
```

```python

import seaborn as sns
#空白畫布
sns.set(style = "whitegrid")
plt.rcParams['font.family']='DFKai-SB' #解决中文乱码
#右邊比例圖
cm = plt.cm.get_cmap('RdYlBu')

fig,ax = plt.subplots(figsize = (12,10))

bubble = ax.scatter(x1, y1, s=z1*30, 
                    c = z1, 
                    cmap = cm, 
                    linewidth = 0.5, 
                    alpha = 0.5)
#X軸
plt.xticks(fontsize=18)
ax.set_xlabel('月份', fontsize = 20,color='green')
ax.set_xticks(range(0,13), minor=False)

#Y軸
plt.yticks(fontsize=18)
ax.set_ylabel('成團率(%)', fontsize=20, color='green')
ax.set_ylim([0, 100])


ax.grid()
fig.colorbar(bubble)


ax.set_title('九州系列', fontsize = 30)

plt.show()
```
